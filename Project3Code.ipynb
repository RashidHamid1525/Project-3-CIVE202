```python
# one trajectory for the time-series plots
trajectory_number = 6
data_subset = ngsim_subset[ngsim_subset['trajectory_number'] == trajectory_number].reset_index(drop=True)

print(data_subset.head())
```

```python
# Create spacing variable
data_subset['spacing'] = data_subset['leader_position(m)'] - data_subset['follower_position(m)']

print(data_subset[['Time', 'leader_position(m)', 'follower_position(m)', 'spacing']].head())
```

```python
# Defining IDM acceleration function

def idm_acceleration(v, s, delta_v, v0, s0, T, a, b, delta):
    s_star = s0 + v * T + (v * delta_v) / (2 * np.sqrt(a * b))
    s_star = max(s_star, s0)
    acceleration = a * (1 - (v / v0) ** delta - (s_star / s) ** 2)
    return acceleration
```

```python
# Data for IDM simulation

time_data = data_subset['Time'].values
leader_position = data_subset['leader_position(m)'].values
leader_speed = data_subset['leader_speed(m/s)'].values
real_follower_position = data_subset['follower_position(m)'].values
real_follower_speed = data_subset['follower_speed(m/s)'].values
real_follower_acc = data_subset['follower_acc(m/s^2)'].values

dt = 0.1

# IDM parameters
v0 = 30
s0 = 2
T = 1.5
a = 1.0
b = 1.5
delta = 4

# Initialize arrays
n_steps = len(time_data)
sim_position = np.zeros(n_steps)
sim_speed = np.zeros(n_steps)
sim_acc = np.zeros(n_steps)

# Start the simulation using the real follower starting values
sim_position[0] = real_follower_position[0]
sim_speed[0] = real_follower_speed[0]
```

```python
# Run the IDM simulation

for i in range(n_steps - 1):
    gap = max(leader_position[i] - sim_position[i], 0.1)
    delta_v = sim_speed[i] - leader_speed[i]

    sim_acc[i] = idm_acceleration(sim_speed[i], gap, delta_v, v0, s0, T, a, b, delta)

    sim_speed[i + 1] = max(sim_speed[i] + sim_acc[i] * dt, 0)
    sim_position[i + 1] = sim_position[i] + sim_speed[i] * dt

print("Simulation complete!")
```

```python
# Compare real and simulated follower position

plt.figure(figsize=(12,6))
plt.plot(time_data, leader_position, label='Leader Position', color='black')
plt.plot(time_data, real_follower_position, label='Real Follower Position', color='blue')
plt.plot(time_data, sim_position, label='Simulated Follower Position', color='red')
plt.xlabel('Time (s)', fontsize=12)
plt.ylabel('Position (m)', fontsize=12)
plt.title('Real and Simulated Vehicle Position vs Time', fontsize=14)
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Compare real and simulated follower speed

plt.figure(figsize=(12,6))
plt.plot(time_data, real_follower_speed, label='Real Follower Speed', color='blue')
plt.plot(time_data, sim_speed, label='Simulated Follower Speed', color='red')
plt.xlabel('Time (s)', fontsize=12)
plt.ylabel('Speed (m/s)', fontsize=12)
plt.title('Real and Simulated Follower Speed vs Time', fontsize=14)
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Compare real and simulated follower acceleration

plt.figure(figsize=(12,6))
plt.plot(time_data, real_follower_acc, label='Real Follower Acceleration', color='blue')
plt.plot(time_data, sim_acc, label='Simulated Follower Acceleration', color='red')
plt.xlabel('Time (s)', fontsize=12)
plt.ylabel('Acceleration (m/s^2)', fontsize=12)
plt.title('Real and Simulated Follower Acceleration vs Time', fontsize=14)
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Compare real and simulated gap distance

real_gap = leader_position - real_follower_position
sim_gap = leader_position - sim_position

plt.figure(figsize=(12,6))
plt.plot(time_data, real_gap, label='Real Gap', color='blue')
plt.plot(time_data, sim_gap, label='Simulated Gap', color='red')
plt.xlabel('Time (s)', fontsize=12)
plt.ylabel('Gap Distance (m)', fontsize=12)
plt.title('Real and Simulated Gap Distance vs Time', fontsize=14)
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
